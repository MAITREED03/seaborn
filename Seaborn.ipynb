Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.



Seaborn is a data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn, along with their typical uses:
1.	Scatter Plot:
•	Use: Display the relationship between two numerical variables.
•	Seaborn Function: sns.scatterplot()
2.	Line Plot:
•	Use: Show the trend of a numerical variable over a continuous interval or time.
•	Seaborn Function: sns.lineplot()
3.	Bar Plot:
•	Use: Display the distribution of a categorical variable across different groups.
•	Seaborn Function: sns.barplot()
4.	Histogram:
•	Use: Visualize the distribution of a single numerical variable.
•	Seaborn Function: sns.histplot()
5.	Box Plot:
•	Use: Display the summary statistics (median, quartiles, and outliers) of a numerical variable for different categories.
•	Seaborn Function: sns.boxplot()



Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.




import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using seaborn
sns.set(style="darkgrid")

# Use the lineplot function to create the plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set plot labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of fmri Dataset")

# Display the plot
plt.show()


Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Plot box plots
plt.figure(figsize=(12, 6))

# Box plot for 'pclass' vs 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'pclass' vs 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.show()



Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.



import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from seaborn
diamonds = sns.load_dataset('diamonds')

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple="stack", kde=True, palette='viridis')

# Adding labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Price with Cut as Hue')

# Display the plot
plt.show()

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plotting the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Passenger Counts by Month and Year')
plt.show()
